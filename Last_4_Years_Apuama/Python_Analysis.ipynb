#### Completing my process for the next three years:

```python
data2023 = pd.read_csv("/home/gusco/Apuama/FSAE2023_Results/FSAE2023_Results - Overall Standings.csv")
data2024 = pd.read_csv("/home/gusco/Apuama/FSAE2024_Results/FSAE2024_Results - Overall Standings.csv")
data2025 = pd.read_csv("/home/gusco/Apuama/FSAE2025_Results/FSAE2025_Results - Overall Standings.csv")

numerical_columns = ['Cost', 'Presentation', 'Design', 'Acceleration', 'Skid Pad', 'Autocross', 'Endurance', 'Efficiency', 'Overall Score']

for col in numerical_columns:
    data2023[col] = pd.to_numeric(data2023[col].str.replace(",", "."), errors="coerce")
    data2024[col] = pd.to_numeric(data2024[col].str.replace(",", "."), errors="coerce")
    data2025[col] = pd.to_numeric(data2025[col].str.replace(",", "."), errors="coerce")

apuama2023 = data2023.iloc[[22]]
apuama2024 = data2024.iloc[[22]]
apuama2025 = data2025.iloc[[18]]
```

```python
numerical_columns = ['Penalties', 'Cost', 'Presentation', 'Design', 'Acceleration', 'Skid Pad', 'Autocross', 'Endurance', 'Efficiency', 'Overall Score']
apuama2022.loc[:, numerical_columns] = apuama2022[numerical_columns].fillna(0).astype(float)
apuama2023.loc[:, numerical_columns] = apuama2023[numerical_columns].fillna(0).astype(float)
apuama2024.loc[:, numerical_columns] = apuama2024[numerical_columns].fillna(0).astype(float)
apuama2025.loc[:, numerical_columns] = apuama2025[numerical_columns].fillna(0).astype(float)
```

```python
apuama2022
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 3))
axes = axes.flatten()
columns = ['Place', 'Overall Score']

years = [2022, 2023, 2024, 2025]
score = [
    apuama2022['Overall Score'].values[0],
    apuama2023['Overall Score'].values[0],
    apuama2024['Overall Score'].values[0],
    apuama2025['Overall Score'].values[0]
]

place = [
    apuama2022['Place'].values[0],
    apuama2023['Place'].values[0],
    apuama2024['Place'].values[0],
    apuama2025['Place'].values[0]
]

for ax, col in zip(axes, columns):
    score = [
        apuama2022[col].values[0],
        apuama2023[col].values[0],
        apuama2024[col].values[0],
        apuama2025[col].values[0]
    ]
    
    ax.plot(years, score, marker='o', color='#005f73', linewidth=2)
    ax.set_title(f'{col} by Year', fontsize=9, fontweight='bold', pad=8)
    ax.set_xlabel('Year', fontsize=8)
    ax.set_ylabel(col, fontsize=8)   # <- já ajusta o label também

    deslocamento = max((max(score) - min(score)) * 0.12, 2)

    for x, y in zip(years, score):
        ax.text(x, y + deslocamento,
                f"{y}", ha='center', fontsize=7)

    # inverter só no gráfico de Place
    if col == "Place":
        ax.invert_yaxis()
        ax.set_ylim(bottom=max(score) + deslocamento, top=min(score) - deslocamento)

    else:
        ax.set_ylim(top=max(score) + deslocamento * 2)


    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(False)
    ax.set_xticks(years)


# Remove extra axles
for ax in axes[len(numerical_columns):]:
    fig.delaxes(ax)

# Adjusts overall spacing
fig.subplots_adjust(hspace=0.5, wspace=0.3)

plt.show()

```

#### Performing the same comparison for each test

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(12, 8))
axes = axes.flatten()

for ax, col in zip(axes, numerical_columns):
    score = [
        apuama2022[col].values[0],
        apuama2023[col].values[0],
        apuama2024[col].values[0],
        apuama2025[col].values[0]
    ]
    
    ax.plot(years, score, marker='o', color='#005f73', linewidth=2)
    ax.set_title(f'{col} Scores by Year', fontsize=9, fontweight='bold', pad=8)
    ax.set_xlabel('Year', fontsize=8)
    ax.set_ylabel('Score', fontsize=8)

    # Proportional displacement + absolute minimum
    deslocamento = max((max(score) - min(score)) * 0.12, 2)
    
    # Adjusting the upper limit to avoid overlapping with the title
    ax.set_ylim(top=max(score) + deslocamento * 2)

    for x, y in zip(years, score):
        ax.text(x, y + deslocamento,
                f"{y}", ha='center', fontsize=7)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(False)
    ax.set_xticks(years)

# Remove extra axles
for ax in axes[len(numerical_columns):]:
    fig.delaxes(ax)

# Adjusts overall spacing
fig.subplots_adjust(hspace=0.5, wspace=0.3)

plt.show()
```
